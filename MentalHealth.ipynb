```python
# data cleaning 
# Dealing with missing data - can be seen using data.info() 
# lets get rid of variables 
check = data.drop(['comments','Timestamp','state'],axis=1)
check.shape
# checking data for missing value
check.isnull().sum().max()
```

```python
# assign default value for each data type
defaultint = 0
defaultString = 'NaN'
defaultfloat = 0.0

stringFeature = ['Gender', 'Country', 'self_employed', 'family_history',
       'treatment', 'work_interfere', 'no_employees', 'remote_work',
       'tech_company', 'benefits', 'care_options', 'wellness_program',
       'seek_help', 'anonymity', 'leave', 'mental_health_consequence',
       'phys_health_consequence', 'coworkers', 'supervisor',
       'mental_health_interview', 'phys_health_interview',
       'mental_vs_physical', 'obs_consequence']

intFeature = ['Age']

floatFeature = []

# clean the NaN's 

for feature in check:
    if feature in intFeature:
        check[feature] = check[feature].fillna(defaultint)
    elif feature in stringFeature:
        check[feature] = check[feature].fillna(defaultString)
    elif feature in floatFeature:
        check[floatFeature] = check[floatFeature].fillna(defaultfloat)
    else:
        print("Error: Feature %s not recognized" % feature)

check.head()
```
